```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
# Population parameters
population_mean = 100
population_stddev = 15
# Sample statistics
sample_mean = 140
sample_size = 30
# Set up hypotheses
# H0: The medication has no effect on intelligence (sample_mean ==population_mean)
# H1: The medication has an effect on intelligence (sample_mean !=population_mean)
# Calculate the z-score
z_score = (sample_mean - population_mean) / (population_stddev /(sample_size ** 0.5))
# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
# Set significance level
alpha = 0.05
# Print results
print("Name: Hitesh Bhanushali")
print("Z-score:", z_score)
print("P-value:", p_value)
# Make a decision
if p_value < alpha:
  print("Reject the null hypothesis. The medication likely affects intelligence.")
else:
  print("Fail to reject the null hypothesis.")
# Generate a range of z-scores for the x-axis
x = np.linspace(-3, 3, 100)
# Calculate the corresponding probability density function (PDF) values for the standard normal distribution
y = stats.norm.pdf(x, 0, 1)
# Plot the standard normal distribution curve
plt.plot(x, y)
plt.title("Standard Normal Distribution Curve")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
# Mark the calculated z-score on the plot
plt.axvline(x=z_score, color='red', linestyle='--', label='Calculated Z-score')
plt.legend()
# Show the plot
plt.show()
```
